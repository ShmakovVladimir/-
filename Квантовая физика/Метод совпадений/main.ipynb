```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau_values = np.array([200, 500, 1000]) * 1e-9
N_values = np.array([2889, 5972, 9070]) / (10 * 60)
N1 = 999913 / (5 * 60) - 16928 / (10 * 60)
N2 = 559913 / (5 * 60) - 7610 / (10 * 60)
n1_full = 999913 / (5 * 60)
n2_full = 559913 / (5 * 60)
```

```python
N_zero_values = []
for N, tau in zip(N_values, tau_values):
    N = N - 2 * tau * n1_full * n2_full
    N_zero = 1.08 * N1 * N2 / (2 * N)
    N_zero /= (3.7 * np.power(10, 10))
    print(round(N_zero * np.power(10, 6), 2),
          round(0.025 * N_zero * np.power(10, 6), 2)
          , "мкКи")
    N_zero_values.append(N_zero)
```

```python
plt.figure(figsize = (12, 5), dpi = 200)

plt.scatter(tau_values,
            np.array(N_zero_values) * np.power(10, 6),
            s = 50)
```
