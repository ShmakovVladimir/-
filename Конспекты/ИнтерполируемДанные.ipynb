# Интерполяция данных в python - конспект
 Для интерполяции будем использовать модуль interpolate библиотеки scipy. В этом модуле находится куча полезных функций, чаще всего для лабораторных работ используется метод interp1d. Он позволяет создавать функции, основанные на входных данных.

### Импортируем библиотеки 
- **matplotlib** для построения графиков
- **numpy** для создания и обработки данных
- из **scipy** импортируем функцию interp1d для интерполяции данных

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
```

### Сгенерируем рандомные данные при помощи numpy
Для этого используем метод `random` из модуля `random`. Метод принимает один аргумент - размер массива(количество рандомных точек, которые нужно сгененрировать)

```python
size = 5 #количество рандомных точек
yRandom = np.random.random(size)
xRandom = np.random.random(size)
```

### Интерполируем рандомные данные
Для интерполяции используем функцию `interp1d`(из библиотеки scipy). Метод возвращает функцию f(x), построенную на основе полученных данных. 

Аргумент `kind` - тип интерполяции:
- linear - линейная интерполяция
- quadratic - квадратичная
- cubic - кубическая

```python
funcCubic = interp1d(xRandom,yRandom,kind = 'cubic') #функция кубической интерполяции
funcQuad = interp1d(xRandom,yRandom,kind = 'quadratic') #функция квадратичной инетрполяции
funcLinear = interp1d(xRandom,yRandom,kind = 'linear') #функция линейной инетрполяции
```

### Построим графики данных функций, и сравним их

Функция, возвращаемая методом `interp1d` может принимать массив numpy в качестве аргумента. Создадим массив точек оси x методом `linspace`, и передадим полученный массив функциям.

```python
xAxes = np.linspace(min(xRandom),max(xRandom),size*1000) 
yCubic = funcCubic(xAxes)
yQuad = funcQuad(xAxes)
yLinear = funcLinear(xAxes)

fig,ax = plt.subplots()
plt.plot(xRandom,yRandom,'o',label = 'Данные') #график начальных данных
plt.plot(xAxes,yLinear,label = 'Линейная интерполяция')
plt.plot(xAxes,yQuad,'--',label = 'Квадратичная интерполяция')
plt.plot(xAxes,yCubic,label = 'Кубическая интерполяция')

ax.legend(loc = 'lower right')

plt.show()
```
