```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
prisma_angle = np.pi / 3
```

```python
data = pd.read_csv("data.csv")
data
```

```python
angles_radians = np.pi * (data.angle + data.minuets / 60 + data.seconds / 3600) / 180 
n = np.sin((prisma_angle + angles_radians) / 2) / np.sin(prisma_angle / 2)
```

```python
plt.figure(figsize = (16, 9))

# plt.errorbar(data.wave_length,
#              n, 
#              xerr = [0 * x for x in data.wave_length],
#              yerr = [0 * x for x in n],
#              fmt = '_')
plt.scatter(data.wave_length, n, c = ['purple', 'blue', 'deepskyblue', 'green', 'yellow', 'yellow', 'red'][::-1])
plt.xlabel("$\lambda$")
plt.ylabel("$n$")
plt.grid()
```
