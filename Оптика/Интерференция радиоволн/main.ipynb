# Интерференция электромагнитных волн миллимитрового диапазона

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import linregress
from scipy.optimize import curve_fit
```

```python
data = pd.read_csv('data.csv')
xErr, IErr = 0.05, 0.5
data.head()
```

```python
plt.figure(figsize = (16, 9), dpi = 500)

plt.errorbar(data.x, 
             data.I,
             xerr = [xErr for _ in data.x],
             yerr = [IErr for _ in data.I],
             fmt = '_',
             label = 'Экспериментальные данные')

plt.plot(data.x,
         data.I,
         linestyle = '--')
plt.grid()
plt.xticks(np.arange(0, np.max(data.x) + 0.5, 0.25))
plt.yticks(np.arange(0, 33, 2.5))
plt.title("Зависимость интенсивности $I$ от координаты зеркала $x$")
plt.ylabel("Сила тока $I \ мкВ$")
plt.xlabel("Координата зеркала $x \ мм$")
plt.legend()
plt.savefig('IntensityVX.svg')
```
