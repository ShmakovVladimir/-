```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
from PIL import Image
import cv2
```

### Эксперимент 1 - измерение периодов решеток

##### Изображения, полученные в ходе эксперимента

```python
filePath = 'experiment_1_photos/'
fileNames = ['1d.jpg', '1u.jpg', '2d.jpg', '2u.jpg', '3d.jpg', '3u.jpg']
upImages, downImages = [], []
```

```python
fig, ax = plt.subplots(2, 3, figsize = (21, 9))

for ind in np.arange(3):
    upImages.append(cv2.imread(filePath + f'{ind + 1}u.jpg'))
    downImages.append(cv2.imread(filePath + f'{ind + 1}d.jpg'))
    ax[0, ind].imshow(upImages[-1])
    ax[1, ind].imshow(downImages[-1])

ax[0, 0].set_ylabel("Верхние сетки")
ax[1, 0].set_ylabel("Нижние сетки")
fig.suptitle('Данные эксперимента')
plt.savefig("ExperimentalData.png")
```

```python
def createMask(img):
    return img[:, :, 1] > 210
```

```python
up_masked, down_masked = [], []
fig, ax = plt.subplots(2, 3, figsize = (21, 9))

for ind in np.arange(3):
    up_masked.append(createMask(upImages[ind]))
    down_masked.append(createMask(downImages[ind]))
    ax[0, ind].imshow(up_masked[-1])
    ax[1, ind].imshow(down_masked[-1])

ax[0, 0].set_ylabel("Верхние сетки")
ax[1, 0].set_ylabel("Нижние сетки")
fig.suptitle('Данные эксперимента')
plt.savefig("ExperimentalData.png")
```
