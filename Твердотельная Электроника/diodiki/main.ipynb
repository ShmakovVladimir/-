```python
import numpy as np
import matplotlib.pyplot as plt
```

### Темновые характеристики

```python
# Заданные точки
U = np.array([0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 1, 1.5, 2.1, 2.5, 3, 4.5, 5, 5.5, 6, 6.4])
I = np.array([0.6, 1.4, 2, 2.6, 3.2, 3.7, 4.2, 4.8, 5.2, 5.8, 6.2, 6.6, 7.2, 7.6, 8.15, 10, 15, 20, 24, 30, 50, 64, 76, 90, 100])


# Аппроксимация прямой по первым нескольким точкам
# Используем первые 10 точек для аппроксимации
mask = U < 1.1
U_approx = U[mask]
I_approx = I[mask]
# Построение графика
plt.figure(figsize = (12, 5), dpi = 200)
plt.scatter(U_approx, I_approx, s = 120, edgecolors = 'black', linewidth = 2, label = 'Экспериментальные данные')
plt.xlabel('Напряжение [В]', fontsize = 16)
plt.ylabel('Сила тока [мкА]', fontsize = 16)
plt.title('Прямая ветвь темновой характеристики', fontsize = 16)
plt.grid(True)


# Полиномиальная аппроксимация первой степени (прямая)
coefficients = np.polyfit(U_approx, I_approx, 1)
poly = np.poly1d(coefficients)

# Вывод уравнения прямой
equation = f"$I = {coefficients[0]:.2f} \cdot U + {coefficients[1]:.2f}$"
print(equation)

# Построение аппроксимирующей прямой
U_line = np.linspace(0, 1.2, 1000)
I_line = poly(U_line)

plt.plot(U_line, I_line, color = 'orange', label = equation, zorder = -1)
# Добавление уравнения прямой на график
#plt.text(0.5, 61, equation, fontsize=13, color='black')

plt.xticks(np.arange(0, 1.2, 0.05))
plt.yticks(np.arange(0, 13, 0.5))
plt.xlim(0, 1.1)
plt.ylim(0, 12)
plt.legend(fontsize = 16)
```

```python
import math
I = [0.6, 1.4, 2, 2.6, 3.2, 3.7, 4.2, 4.8, 5.2, 5.8, 6.2, 6.6, 7.2, 7.6, 8.15, 10, 15, 20, 24, 30, 50, 64, 76, 90, 100]
I = [math.log(x) for x in I]
U = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 1, 1.5, 2.1, 2.5, 3, 4.5, 5, 5.5, 6, 6.4]


# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(U, I, 'o')
plt.xlabel('U, В')
plt.ylabel('ln(I)')
plt.title('График завиимости ln(I)=f(U) для прямой ветви темнового тока')
plt.grid(True)

# Аппроксимация прямой по первым нескольким точкам
# Используем первые 10 точек для аппроксимации
U_approx = U[:10]
I_approx = I[:10]

# Полиномиальная аппроксимация первой степени (прямая)
coefficients = np.polyfit(U_approx, I_approx, 1)
poly = np.poly1d(coefficients)

# Вывод уравнения прямой
equation = f"I = {coefficients[0]:.2f} * U + {coefficients[1]:.2f}"
print(equation)

# Построение аппроксимирующей прямой
U_line = np.linspace(min(U), 1, 50)
I_line = poly(U_line)

plt.plot(U_line, I_line, 'r--')

# Добавление уравнения прямой на график
plt.text(0.1, 4.5, equation, fontsize=13, color='black')


U_approx = U[10:25]
I_approx = I[10:25]

# Полиномиальная аппроксимация первой степени (прямая)
coefficients = np.polyfit(U_approx, I_approx, 1)
poly = np.poly1d(coefficients)

# Вывод уравнения прямой
equation = f"I = {coefficients[0]:.2f} * U + {coefficients[1]:.2f}"
print(equation)

# Построение аппроксимирующей прямой
U_line = np.linspace(0, max(U), 50)
I_line = poly(U_line)

plt.plot(U_line, I_line, 'r--')

# Добавление уравнения прямой на график
plt.text(2.5, 2.5, equation, fontsize=13, color='black')
plt.show()
```

### Световая характеристика

```python
from scipy.optimize import curve_fit

# Данные
x = np.array([0, 0.14, 0.27, 0.35, 0.5, 0.7, 1.4, 2, 1.7, 1.5, 2.1]) * 3
y = np.array([6.9, 7, 7, 7, 7, 7, 6.9, 3, 6.2, 6.7, 0]) * 2

# Функция для аппроксимации
def func(x, A, b, c, O):
    return A * np.exp((x - b) / c) + O

# Начальные значения параметров
initial_guess = [1, 1, 1, 1]

# Оптимизация параметров
params, _ = curve_fit(func, x, y, p0=initial_guess)

# Параметры
A, b, c, O = params

# Построение графика данных
plt.figure(figsize = (12, 5), dpi = 200)
plt.scatter(x, y, label = 'Экспериментальные данные', zorder = 10, s = 70, edgecolors='black', linewidths=2)

# Построение аппроксимирующей функции
x_fit = np.linspace(0, 10, 1000)
y_fit = func(x_fit, A, b, c, O)
plt.plot(x_fit, y_fit, color = 'orange',
         label=r'$y = A \cdot \exp(\frac{x - b}{c}) + O$' + '\nA = {}, b = {}, c = {}, O = {}'.format(*[p.round(4) for p in params], zorder = 1))
plt.grid(True)
# Добавление легенды
plt.legend(fontsize = 16)

# Подписи осей

plt.xlabel('Напряжение [В]', fontsize = 16)
plt.ylabel('Сила тока [мА]', fontsize = 16)
plt.xticks(np.arange(0, 7, 0.25), rotation = 45)
plt.xlim(-0.1, 6.5)
plt.yticks(np.arange(0, 15, 0.5))
plt.ylim(-0.1, 14.5)
# Заголовок графика
plt.title('Световая характеристика без фильтра', fontsize = 16)
# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Новые данные
x = np.array([0.5, 0.9, 1.2, 1.45, 1.7, 1.9, 1.98]) * 3
y = np.array([3.5, 3.5, 3.3, 3, 2.5, 1.1, 0]) * 2


# Функция для аппроксимации
def func(x, A, b, c, O):
    return A * np.exp((x - b) / c) + O

# Начальные значения параметров
initial_guess = [0, 4, 1, 10]

# Оптимизация параметров
params, _ = curve_fit(func, x, y, p0=initial_guess)

# Параметры
A, b, c, O = params

# Построение графика данных
plt.figure(figsize = (12, 5), dpi = 200)
plt.scatter(x, y, label = 'Экспериментальные данные', zorder = 10, s = 70, edgecolors='black', linewidths=2)

# Построение аппроксимирующей функции
x_fit = np.linspace(0, 7, 1000)
y_fit = func(x_fit, A, b, c, O)
plt.plot(x_fit, y_fit, color = 'orange',
         label=r'$y = A \cdot \exp(\frac{x - b}{c}) + O$' + '\nA = {}, b = {}, c = {}, O = {}'.format(*[p.round(4) for p in params]))
plt.grid(True)
# Добавление легенды
plt.legend(fontsize = 16)

# Подписи осей
plt.axvline(b)

plt.xlabel('Напряжение [В]', fontsize = 16)
plt.ylabel('Сила тока [мА]', fontsize = 16)

plt.xticks(np.arange(0, 7, 0.25), rotation = 45)
plt.xlim(-0.1, 6.5)
plt.yticks(np.arange(0, 15, 0.5))
plt.ylim(-0.1, 7.5)
# Заголовок графика
plt.title('Световая характеристика с темным фильтром', fontsize = 16)

# Отображение графика
plt.show()
```
