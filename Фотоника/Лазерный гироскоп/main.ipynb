```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import cm
```

```python
names = [r'pump_minus_1.2_voltage_92.3.csv', r'pump_minus_1.2_voltage_92.3_back.csv', r'pump_minus_1.2_voltage_92.3_hui_poimi.csv']
```

```python
plt.figure(figsize = (12, 5), dpi = 200)
for i, name in enumerate(names):
    data = pd.read_csv(name)
    plt.scatter(data.V, data.deltaNu, color = cm.rainbow(i / len(names)))
```
