```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
```

```python
data = pd.read_csv("data.csv")
data
```

```python
xInt = interp1d(data.x,data.r_L,kind = "quadratic")
fig,ax = plt.sublots(figsize = (16,9))
plt.errorbar(
    x = data.x,
    y = data.r_L,
    xerr = 0.05*data.x,
    yerr = 0.025*data.r_L,
    fmt = '.'
)

plt.plot(np.linspace(5.01,39.99,1000),
         xInt(np.linspace(5.01,39.99,1000)))

plt.xlabel("Положение сердечника мм")
plt.show()
```
