```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('vah.csv')

plt.figure(figsize = (12, 5), dpi = 200)
plt.scatter(data.U[:8], data.I[:8], label = 'Эксперимент 1')
plt.scatter(data.U[8:], data.I[8:], label = 'Эксперимент 2')
plt.xlabel("Напряжение [В]", fontsize = 16)
plt.ylabel("Сила тока [мка]", fontsize = 16)
plt.grid(ls = ':')
plt.legend()
```
